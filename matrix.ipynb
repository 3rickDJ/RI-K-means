# pesado
$w = tf \cdot idf\$

$w :=$ pesado de cada elemento en la tabla

$tf :=$ es el elemento en la columna y fila

$idf = log\frac{N}{n_k} $

N = total de documentos

$n_k$ = numero de documentos donde aparece el stem


```python
N = df.shape[0] # N
```

```python
stems = df.columns[1:]
results = pd.DataFrame([], columns=df.columns)
def is_present(column, number=0):
    return (column > number).sum()
n_k = df.drop(columns='name').apply(is_present)

id_k = np.log2(N/n_k) 
id_k.hasnans
```

```python
tf_idf = df[stems].mul(id_k, axis=1)
tf_idf.tail()
```
